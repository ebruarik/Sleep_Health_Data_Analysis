Uyku Sağlığı ve Yaşam Tarzı Analizi (Veri Temizleme Projesi)

Bu proje, bireylerin yaşam tarzı alışkanlıklarının uyku kalitesi üzerindeki etkilerini 
incelemek amacıyla hazırlanan veri setinin temizlenmesi ve analize hazır hale getirilmesi sürecini kapsar.

1: Kütüphanelerin Yüklenmesi ve Verinin Okunması

Analiz sürecinde kullanacağımız temel Python kütüphanelerini yüklüyoruz ve veri setimizi projemize dahil ediyoruz.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')

df = pd.read_csv('Sleep_health_and_lifestyle_dataset.csv')

df.head()
```

```python
df.info()
```

2: Eksik Veri Yönetimi

Veri setindeki "Sleep Disorder" sütununda yer alan boş değerleri analiz ediyoruz.
Bu boşluklar birer hata değil, uyku bozukluğu olmayan bireyleri temsil etmektedir. 
Bu veriler 'None' etiketiyle doldurularak veri seti bütünlüğü sağlanacaktır.

```python
print("eksik değerler:" , df.isnull().sum())

df['Sleep Disorder'] = df['Sleep Disorder'].fillna('None')

print("doldurma sonrası  eksik değerler:" , df.isnull().sum())
df.head(10)
```

3: Veri Ayrıştırma

Metin formatında olan 'Blood Pressure' (tansiyon ) verisi matematiksel analize uygun hale getirilmesi için '/'
karakterinden bölünerek büyük(Systolic_BP) ve küçük(Diastolic_BP) olmak üzere iki ayrı sayısal sütuna dönüştürülecek.
    

```python
df[['Systolic_BP' , 'Diastolic_BP']] = df['Blood Pressure'].str.split('/', expand = True).astype(int)

df[['Blood Pressure', 'Systolic_BP' , 'Diastolic_BP']].head()
```

4: Keşifçi Veri Analizi (EDA) ve Görselleştirme

Bu aşamada, temizlediğimiz veriler arasındaki ilişkileri görselleştirerek anlamlı içgörüler elde ediyoruz. 
Özellikle meslek grupları, stres seviyeleri ve yeni oluşturduğumuz tansiyon sütunları arasındaki bağlantıları inceleyeceğiz.

```python
plt.figure(figsize=(12, 6))
sns.barplot(x='Occupation', y='Stress Level', data=df, palette='viridis', errorbar=None)
plt.title('Mesleklere Göre Ortalama Stres Seviyesi')
plt.xticks(rotation=45) 
plt.show()
```

Satış Temsilcileri (Sales Representatives) en yüksek stres seviyesine (8) sahip meslek grubudur.
Mühendisler ve Yöneticiler, Doktorlara kıyasla daha düşük stres seviyeleri raporlamıştır.
Bu yüksek stres seviyesinin, dün ayrıştırdığımız Systolic_BP (Büyük Tansiyon) verilerini yukarı çekip çekmediği 
bir sonraki korelasyon analizimizde incelenecektir.

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
sns.regplot(x='Stress Level', y='Systolic_BP', data=df, scatter_kws={'alpha':0.5}, line_kws={'color':'red'})
plt.title('Stres Seviyesi ve Büyük Tansiyon (Systolic) İlişkisi')
plt.xlabel('Stres Seviyesi')
plt.ylabel('Büyük Tansiyon')
plt.show()
```

Stres ve Büyük Tansiyon (Systolic_BP) arasında yaptığımız regresyon analizinde, iki değişken arasında anlamlı bir korelasyon saptanmamıştır. 
Stres seviyesi artsa bile tansiyon değerleri benzer bir dağılım sergilemektedir.
Tansiyon üzerinde stresin yanı sıra; 'BMI Category' (Kilo), 'Age' (Yaş) veya 'Physical Activity Level' gibi diğer faktörlerin 
daha baskın bir etkisi olabilir.

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x='BMI Category', y='Systolic_BP', data=df, palette='Set2')
plt.title('Vücut Kitle Endeksi (BMI) ve Büyük Tansiyon İlişkisi')
plt.show()
```

Tansiyonu stres yükseltmiyorsa ne yükseltiyor? 
(BMI (Kilo durumu) ile olan ilişkisine bakıldı çünkü kilonun tansiyonla ilişkisi genellikle çok daha güçlüdür.)
Yukarıdaki grafikte de görüldüğü üzere stres faktörünün aksine, BMI kategorisinin tansiyon değerleri üzerinde doğrudan ve güçlü bir etkisi olduğu gözlemlenmiştir. Overweight kategorisindeki bireylerin büyük tansiyon ortalaması belirgin şekilde yüksektir. 
Normal kilolu bireylerde tansiyon değerleri daha düşük ve dengeli bir dağılım sergilemektedir.
Bu bulgu, uyku sağlığı ve tansiyon üzerinde kilo yönetiminin stresten daha öncelikli bir risk faktörü olabileceğini göstermektedir.
